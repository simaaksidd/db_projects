This query uses an inner join that joins the actor and film actor tables that looks for actor_id's greater than 10 and less than 30 ordered by their last name and displays everything

```python
!psql -c "select * from pagila.actor a inner join pagila.film_actor fa on a.actor_id=fa.actor_id where a.actor_id>10 and a.actor_id<30 order by a.last_name"
```

This query uses an inner join that joins the rental, payment, and customer tables that looks for amount less than 5 ordered by payment date and displays customer first name, last name, and amount

```python
!psql -c "select c.first_name, c.last_name, p.amount from pagila.rental r inner join pagila.payment p on r.customer_id=p.customer_id inner join pagila.customer c on r.customer_id=c.customer_id where p.amount<5 order by c.last_name limit 20"
```

This query uses an inner join that joins the store, inventory, and film tables that looks for film_title that is ARMY fLINTSTONES orderd by rental duration and displays store_id and inventory id

```python
!psql -c "select s.store_id, i.inventory_id from pagila.store s inner join pagila.inventory i on s.store_id=i.store_id inner join pagila.film f on i.film_id=f.film_id where f.title='ARMY FLINTSTONES' order by f.rental_duration limit 20"
```

This query uses an outer join that left joins the store and the customer tables that look for stores with customers that input their email ordered by store id and displays customer emails for a special promotion

```python
!psql -c "select c.email from pagila.store s left outer join pagila.customer c on s.store_id=c.store_id where c.email is not null order by s.store_id limit 20"
```

This query uses an outer join that left joins the language and the film tables that look for films in English ordered by film name and displays film name (we wanted to do any other language but turns out every film is in english)

```python
!psql -c "select f.title from pagila.language l left outer join pagila.film f on l.language_id=f.language_id where l.name='English' order by f.title limit 20"
```

This query uses an outer join that left joins the rental and the customer tables that look for return dates that are null (not returned) ordered by last update and displays first name and email 

```python
!psql -c "select c.first_name, c.last_name, c.email, r.rental_date from pagila.rental r left outer join pagila.customer c on r.customer_id=c.customer_id where r.return_date is null order by r.rental_date limit 20"
```

This query uses an self join that looks at the actor table and tries to find actors who have the same first name, but not the same person.

```python
!psql -c "select a1.first_name, a1.last_name from pagila.actor a1 inner join pagila.actor a2 on a1.first_name=a2.first_name and a1.actor_id!=a2.actor_id order by a1.first_name limit 20"
```
